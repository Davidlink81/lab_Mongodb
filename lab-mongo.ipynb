# Advanced Querying Mongo

Importing libraries and setting up connection with "companies" collection. As in class, call it `c`

```python
from pymongo import MongoClient
client = MongoClient("localhost:27017")
```

```python
#Your code here
client.list_database_names()
```

```python
db=client["Ironhack"]
```

```python
c=db.get_collection("Companies")
```

### 1. All the companies whose name match 'Babelgum'. Retrieve only their `name` field.

```python
# Your Code
list(c.find({"name": "Babelgum"},{"_id":0, "name":1}))
```

### 2. All the companies that have more than 5000 employees. Limit the search to 20 companies and sort them by **number of employees**.

```python
# Your Code
list(c.find({"number_of_employees":{"$gt": 5000}},{"_id":0, "name":1},).limit(20))
```

### 3. All the companies founded between 2000 and 2005, both years included. Retrieve only the `name` and `founded_year` fields.

```python
# Your Code
list(c.find({"founded_year":{"$gte": 2000, "$lte": 2005}},{"_id":0, "name":1, "founded_year":1}))
```

### 4. All the companies that had a Valuation Amount of more than 100.000.000 and have been founded before 2010. Retrieve only the `name` and `ipo` fields.

```python
# Your Code
c.find_one()
```

```python
list(c.find({"founded_year":{"$lt": 2010}, "ipo.valuation_amount":{"$gt": 100000000}},{"_id":0, "name":1, "founded_year":1,"ipo.valuation_amount":1}))
```

### 5. All the companies that have less than 1000 employees and have been founded before 2005. Order them by the number of employees and limit the search to 10 companies.

```python
# Your Code
list(c.find({"number_of_employees":{"$gt": 1000},"founded_year":{"$lt": 2005}},{"_id":0, "founded_year":1, "number_of_employees":1, "name":1}).sort("number_of_employees", -1).limit(10))
```

### 6. All the companies that don't include the `partners` field.

```python
# Your Code
list(c.find({"partners":{"$exists": False}},{"_id":0, "name":1, "partners":1}))
```

### 7. All the companies that have a null type of value on the `category_code` field.

```python
# Your Code
list(c.find({"category_code": None},{"_id":0, "name":1, "category_code":1}))
```

### 8. All the companies that have at least 100 employees but less than 1000. Retrieve only the `name` and `number of employees` fields.

```python
# Your Code
list(c.find({"number_of_employees":{"$gte": 100, "$lt": 1000}},{"_id":0, "name":1, "number_of_employees":1}))
```

### 9. Order all the companies by their IPO price in a descending order.

```python
# Your Code
list(c.find({},{"_id":0, "name":1, "ipo.price":1}).sort("ipo.price", 1))
```

### 10. Retrieve the 10 companies with more employees, order by the `number of employees`

```python
# Your Code
list(c.find({},{"_id":0, "number_of_employees":1, "name":1}).sort("number_of_employees", -1).limit(10))
```

### 11. All the companies founded on the second semester of the year. Limit your search to 1000 companies.

```python
# Your Code
list(c.find({"founded_month":{"$gte": 6, "$lte": 12}},{"_id":0, "name":1, "founded_month":1}).limit(1000))
```

### 12. All the companies founded before 2000 that have an acquisition amount of more than 10.000.00

```python
list(c.find({"founded_year":{"$lt":2000}, "acquisition.amount_amount":{"$gt":10000}},{"_id":0, "name":1, "founded_year":1, "acquisition.amount_amount":1 }))
```

### 13. All the companies that have been acquired after 2010, order by the acquisition amount, and retrieve only their `name` and `acquisition` field.

```python
# Your Code
list(c.find({"acquisition.acquired_year":{"$lt":2010}}, {"_id":0, "name":1, "acquisition.price_amount":1}).sort("acquisition.price_amount",1))
```

### 14. Order the companies by their `founded year`, retrieving only their `name` and `founded year`.

```python
# Your Code
list(c.find({},{"_id":0, "name":1,"founded_year":1}).sort("founded_year",-1))
```

### 15. All the companies that have been founded on the first seven days of the month, including the seventh. Sort them by their `acquisition price` in a descending order. Limit the search to 10 documents.

```python
# Your Code
list(c.find({"founded_day":{"$gte":0},"founded_day":{"$lte":7}},{"_id":0, "name":1,"founded_day":1}).sort("acquisition.price_amount",-1).limit(10))
```

### 16. All the companies on the 'web' `category` that have more than 4000 employees. Sort them by the amount of employees in ascending order.

```python
# Your Code
list(c.find({"category_code": "web","number_of_employees":{"$gte":4000}},{"_id":0, "name":1,"number_of_employees":1}).sort("number_of_employees",1))
```

### 17. All the companies whose acquisition amount is more than 10.000.000, and currency is 'EUR'.

```python
# Your Code
list(c.find({"acquisition.price_amount":{"$gt":10000000}, "acquisition.price_currency_code":"EUR"},{"_id":0, "name":1, "acquisition.price_amount":1, "acquisition.price_currency_code":1}))
```

### 18. All the companies that have been acquired on the first trimester of the year. Limit the search to 10 companies, and retrieve only their `name` and `acquisition` fields.

```python
# Your Code
list(c.find({"acquisition.acquired_month": {"$in":[1,2,3]}},{"_id":0, "name":1,"acquisition":1}).limit(10))
```
